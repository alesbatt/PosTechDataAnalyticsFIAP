```python
receita_por_pais = dados_exportacao.groupby('Destino')['Valor'].sum().sort_values(ascending = False)
receita_por_pais
```

```python
#existem 11 países da qual não exportamos nos últimos 15 anos
receita_por_pais[100:128]
```

```python
#dados_exportacao[dados_exportacao['Destino'] == "Anguilla"]
```

```python
volume_por_pais = dados_exportacao.groupby('Destino')['Quantidade'].sum().sort_values(ascending = False)
volume_por_pais
```

```python
volume_por_pais[100:128]
```

```python
plt.figure(figsize=(10, 7))
axis = receita_por_pais[:10].plot(kind='barh', color = 'purple')
axis.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.1f}"))
plt.title('TOP 10 - Países com maior receita de exportação', fontsize = 16)
plt.xlabel('Receita de exportação')
plt.ylabel('Destino')
plt.show()
```

```python
plt.figure(figsize=(10, 7))
axis = volume_por_pais[:10].plot(kind='barh', color = 'purple')
axis.xaxis.set_major_formatter(ticker.StrMethodFormatter("{x:,.1f}"))
plt.title('TOP 10 - Países com maior volume de exportação', fontsize = 16)
plt.xlabel('Volume exportado')
plt.ylabel('Destino')
plt.show()
```
